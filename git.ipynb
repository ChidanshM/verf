```python
import subprocess

# Step 1: Get the list of deleted files
deleted_files = subprocess.check_output('git ls-files --deleted', shell=True).decode('utf-8').splitlines()

# Step 2: Remove each file from Git index
if deleted_files:
    for file in deleted_files:
        print(f"Removing {file}")
        subprocess.run(['git', 'rm', file])
else:
    print("No deleted files to remove.")
```
